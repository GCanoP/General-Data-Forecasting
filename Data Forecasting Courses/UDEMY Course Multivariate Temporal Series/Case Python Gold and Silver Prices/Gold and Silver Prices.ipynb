## <font color='Cornflowerblue'> Stationarity Test. </font>
To use a Vectorial Autoregressive Model, the data must be stationary. Then a Grainger Test is applied.

```python
# Create a Dickey-Fuller test function. 
def augmented_dickey_fuller_statistics(time_series):
    """ 
    Return the Dickey-Fuller test for a time serie.
    Dickey-Fuller test provide a wide range of metrics. However, for the purpose of this
    analysis, just the second metric i.e p-value is returned.
    
    Args:
        time_series (pandas.core.frame.DataFrame) : time series. 
    
    Returns:    
    
    """
    result = sts.adfuller(time_series.values)
    return print('p-value : {result}'.format(result = result[1]))
```

```python
# Print Dickey-Fuller test results. 
# Nule hypothesis :          The time serie is not stationary. 
# Alternative hypothesis :   The time serie is stationary.
# Significanve level 0.05. 
# If a higher value is reported nule hypothesis is not rejected. 

print('Augmented Dickey-Fuller test results.')
print('Temporal Serie - Gold Price')
augmented_dickey_fuller_statistics(X_train['Gold'])
print('Temporal Serie - Silver Price')
augmented_dickey_fuller_statistics(X_train['Silver'])
print('Temporal Serie - Oil Price')
augmented_dickey_fuller_statistics(X_train['Oil'])
print('Temporal Serie - Treasury Bonds Price')
augmented_dickey_fuller_statistics(X_train['Treasury'])
print('Temporal Serie - S&P 500 Price')
augmented_dickey_fuller_statistics(X_train['SP500'])
print('Temporal Serie - USD/EUR Price')
augmented_dickey_fuller_statistics(X_train['USD/EUR'])
```

## <font color='Cornflowerblue'> Create the VAR Model. </font>

```python
# Create the simple VAR Model. 
model = VAR(X_train_transformed)
```

```python
# Select the order of the model. 
model_sel = model.select_order(15)
model_sel.summary()
```

```python
# Fitting the model. 
res = model.fit(maxlags=15, ic='aic')
res.summary()
```

## <font color='goldenrod'> Causality Granger Matrix. </font>

```python
# Create a Granger Causality Matrix. 

max_lags = 15
test = 'ssr_chi2test'

def granger_causality_matrix(X_train_transformed, variables, test, verbose = False):
    """
    Return the Granger Causality Matrix. 
    
    Args:
        X_train_transformed (pandas.core.frame.DataFrame) : time series.
        variables (string) : time series name columns. 
        test (string) : statistical test applied. 
        verbose (Boolean) : false by default. 
    
    Return:
        dataset (pandas.core.frame.DataFrame) :  granger causality matrix. 
    """
    dataset = pd.DataFrame(np.zeros((len(variables), len(variables))), columns = variables, index = variables)
    for c in dataset.columns:
        for r in dataset.index:
            test_result = grangercausalitytests(X_train_transformed[[r, c]], maxlag=max_lags, verbose = False)
            p_values = [round(test_result[i+1][0][test][1], 4) for i in range(max_lags)]
            if verbose:
                print(f'Y = {r}, X = {c}, P Values = {p_values}')
            min_p_value = np.min(p_values)
            dataset.loc[r, c] = min_p_value
    dataset.columns = [var + '_x' for var in variables]
    dataset.index = [var + '_y' for var in variables]
    return dataset
```

```python
granger_causality_matrix(X_train_transformed = X_train_transformed, 
                         variables = X_train_transformed.columns, 
                         test = test, 
                         verbose= False)
```

## <font color='Cornflowerblue'> Make Predictions. </font>

```python
# Obtain the order of the model. 
lag_order = res.k_ar
print('Order of the model : {lag_order}'.format(lag_order = lag_order))
```

```python
# Input data for predictions. 
input_data = X_train_transformed.values[-lag_order:]
# Predictions. 
pred = res.forecast(y = input_data, steps = n_obs)
pred = (pd.DataFrame(pred, index=X_test.index, columns=X_test.columns + '_pred'))
print('Predictions:')
pred
```

```python
# Plot predictions
plt.figure(figsize = (12, 10))
res.plot_forecast(15)
plt.tight_layout(h_pad= 1.15)
plt.show()
```

## <font color='Cornflowerblue'> Inverse Transform. </font>

```python
# Inverse transformation. 
def invert_transform(X_train, pred):
    """
    Return the inverse diff transformation. 
    Args:
        X_train (pd.core.DataFrame)
        pred (pd.core.DataFrame)
    Return:
        forecast (pd.core.DataFrame)
    """
    forecast = pred.copy()
    columns = X_train.columns
    for col in columns:
        forecast[str(col) + '_pred'] = X_train[col].iloc[-1] + forecast[str(col) + '_pred'].cumsum()
    return forecast
```

```python
output = invert_transform(X_train, pred)
output
```

```python
# Gold Prediction. 
plt.figure(figsize = (9, 7))
plt.plot(output.iloc[:,0])
plt.title('Gold Predictions')
plt.grid()
plt.show()
```

## <font color='Cornflowerblue'> Evaluate Predictions. </font>

```python
combine = pd.concat([output['Gold_pred'], X_test['Gold']], axis = 1)
combine = combine.round(decimals = 2)
combine = combine.reset_index()
combine = combine.sort_values(by = 'Date', ascending = False)
```

```python
# Plot predictions
combine['Gold'].plot()
combine['Gold_pred'].plot()
plt.legend()
plt.grid()
```

```python
# Metrics for gold predictions. 
from sklearn.metrics import mean_absolute_error, mean_squared_error
print('Mean absolute error : ', mean_absolute_error(combine['Gold'].values, combine['Gold_pred'].values))
print('Root mean squared error : ', np.sqrt(mean_squared_error(combine['Gold'].values, combine['Gold_pred'].values)))
```
